```python
"""
#1
By Keon Hee Kim
"""

import numpy as np
import tables as tb
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

particles_h5file = tb.open_file('particles_0000.h5', 'r')
particles = particles_h5file.root.particles.read()
x = particles[:,0] * 1e3 #[mm]
xp = particles[:,1] * 1e3 #[mrad]
y = particles[:,2] * 1e3 #[mm]
yp = particles[:,3] * 1e3 #[mrad]
z = particles[:,4] * 1e3 #[mm]
zp = particles[:,5] * 1e3 #[mrad]
particles_numbers = particles[:,6]
particles_h5file.close()

#Plot horizontal phase space
fig1 = plt.figure(figsize = (10,10))
fig1.suptitle('Horizontal Phase Space', size=20)
gs1 = GridSpec(4, 4)

ax1_scatter = fig1.add_subplot(gs1[1:4, 0:3])
ax1_scatter.set_xlabel('x [mm]', fontsize=15)
ax1_scatter.set_ylabel('xp [mrad]', fontsize=15)
ax1_hist_x = fig1.add_subplot(gs1[0, 0:3])
ax1_hist_xp = fig1.add_subplot(gs1[1:4, 3])

ax1_scatter.scatter(x, xp, s=5)

ax1_hist_x.hist(x, histtype="step", bins=100)
ax1_hist_xp.hist(xp, histtype="step", bins=100, orientation = 'horizontal')

fig1.tight_layout()
fig1.subplots_adjust(hspace=0.4,wspace=0.4)

#Plot vertical phase space
fig2 = plt.figure(figsize = (10,10))
fig2.suptitle('Vertical Phase Space', size=20)
gs2 = GridSpec(4, 4)

ax2_scatter = fig2.add_subplot(gs2[1:4, 0:3])
ax2_scatter.set_xlabel('y [mm]', fontsize=15)
ax2_scatter.set_ylabel('yp [mrad]', fontsize=15)
ax2_hist_y = fig2.add_subplot(gs2[0, 0:3])
ax2_hist_yp = fig2.add_subplot(gs2[1:4, 3])

ax2_scatter.scatter(y, yp, s=5)

ax2_hist_y.hist(y, histtype="step", bins=100)
ax2_hist_yp.hist(yp, histtype="step", bins=100, orientation = 'horizontal')

fig2.tight_layout()
fig2.subplots_adjust(hspace=0.4,wspace=0.4)

#Plot cross-section phase space
fig3 = plt.figure(figsize = (10,10))
fig3.suptitle('Cross-section Phase Space', size=20)
gs3 = GridSpec(4, 4)

ax3_scatter = fig3.add_subplot(gs3[1:4, 0:3])
ax3_scatter.set_xlabel('x [mm]', fontsize=15)
ax3_scatter.set_ylabel('y [mm]', fontsize=15)
ax3_hist_x = fig3.add_subplot(gs3[0, 0:3])
ax3_hist_y = fig3.add_subplot(gs3[1:4, 3])

ax3_scatter.scatter(x, y, s=5)

ax3_hist_x.hist(x, histtype="step", bins=100)
ax3_hist_y.hist(y, histtype="step", bins=100, orientation = 'horizontal')

fig3.tight_layout()
fig3.subplots_adjust(hspace=0.4,wspace=0.4)
```
